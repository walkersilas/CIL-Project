```python
import os
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
from google.colab import drive
from google.colab import files

drive.mount('/content/gdrive')
DATA_PATH = "/content/gdrive/My Drive/ETH/Computational Intelligence Lab/CIL-Project/experiments/figures/learning_curve/"
```

```python
with open(os.path.join(DATA_PATH, "gnn_ncf_training_loss.json")) as f:
    gnn_ncf_training_json = json.load(f)

with open(os.path.join(DATA_PATH, "gnn_ncf_validation_loss.json")) as f:
    gnn_ncf_validation_json = json.load(f)

with open(os.path.join(DATA_PATH, "reinforced_gnn_ncf_slopeone_training_loss.json")) as f:
    rgnn_training_json = json.load(f)

with open(os.path.join(DATA_PATH, "reinforced_gnn_ncf_slopeone_validation_loss.json")) as f:
    rgnn_validation_json = json.load(f)
```

```python
len(gnn_ncf_training_json[0]["y"])
```

```python
plt.rc('axes', titlesize=25)  # fontsize of the axes title
plt.rc('axes', labelsize=32)  # fontsize of the x and y labels
plt.rc('xtick', labelsize=24)  # fontsize of the tick labels
plt.rc('ytick', labelsize=24)  # fontsize of the tick labels
plt.rc('legend', fontsize=50)  # legend fontsize

subplots = (1, 1)
fig, ax = plt.subplots(subplots[0], subplots[0], figsize=(10, 8))

colors = ["black", "blue", "red", "lime"]
markers = ["v", "s", "o", "d"]
labels = ["RGNN training loss", "RGNN validation loss", "GNN+NCF training loss", "GNN+NCF validation loss"]

ax.plot(np.mean(np.array(gnn_ncf_training_json[0]["x"][4:]).reshape(-1, 7), axis=1),
                np.mean(np.array(gnn_ncf_training_json[0]["y"][4:]).reshape(-1, 7), axis=1),
                marker=markers[2],
                markersize=9,
                linewidth=3,
                mew=3,
                #markeredgecolor="black",
                fillstyle="none",
                color=colors[2],
                label=labels[2])

ax.plot(gnn_ncf_validation_json[0]["x"],
                gnn_ncf_validation_json[0]["y"],
                marker=markers[3],
                markersize=9,
                linewidth=3,
                mew=3,
                #markeredgecolor="black",
                fillstyle="none",
                color=colors[3],
                label=labels[3])

ax.plot(np.mean(np.array(rgnn_training_json[0]["x"][4:]).reshape(-1, 7), axis=1),
                np.mean(np.array(rgnn_training_json[0]["y"][4:]).reshape(-1, 7), axis=1),
                marker=markers[0],
                markersize=9,
                linewidth=3,
                mew=3,
                #markeredgecolor="black",
                fillstyle="none",
                color=colors[0],
                label=labels[0])

ax.plot(rgnn_validation_json[0]["x"],
                rgnn_validation_json[0]["y"],
                marker=markers[1],
                markersize=9,
                linewidth=3,
                mew=3,
                #markeredgecolor="black",
                fillstyle="none",
                color=colors[1],
                label=labels[1])

#ax.set_title(name)
ax.set_xlabel("Step")
ax.set_ylabel("MSE loss")
ax.legend(loc="upper right", fontsize=16)
fig.savefig("./learning_curve.png")
```

```python
plt.rc('axes', titlesize=25)  # fontsize of the axes title
plt.rc('axes', labelsize=32)  # fontsize of the x and y labels
plt.rc('xtick', labelsize=24)  # fontsize of the tick labels
plt.rc('ytick', labelsize=24)  # fontsize of the tick labels
plt.rc('legend', fontsize=50)  # legend fontsize

subplots = (1, 1)
fig, ax = plt.subplots(subplots[0], subplots[0], figsize=(10, 8))

colors = ["black", "blue", "red", "lime"]
markers = ["v", "s", "o", "d"]
labels = ["RGNN training loss", "RGNN validation loss", "GNN+NCF training loss", "GNN+NCF validation loss"]


ax.plot(gnn_ncf_validation_json[0]["x"],
                gnn_ncf_validation_json[0]["y"],
                marker=markers[3],
                markersize=9,
                linewidth=3,
                mew=3,
                #markeredgecolor="black",
                fillstyle="none",
                color=colors[3],
                label=labels[3])

ax.plot(rgnn_validation_json[0]["x"],
                rgnn_validation_json[0]["y"],
                marker=markers[1],
                markersize=9,
                linewidth=3,
                mew=3,
                #markeredgecolor="black",
                fillstyle="none",
                color=colors[1],
                label=labels[1])

#ax.set_title(name)
ax.set_xlabel("Step")
ax.set_ylabel("MSE loss")
ax.legend(loc="upper right", fontsize=16)
fig.savefig("./learning_curve_validation.png")
```
